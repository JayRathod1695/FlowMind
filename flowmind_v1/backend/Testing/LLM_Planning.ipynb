```python
def send_user_request(user_request, model="deepseek-ai/deepseek-v3.1-terminus", stream=True):
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "Speak in english only"},
            {"role": "user", "content": user_request},
        ],
        temperature=0.2,
        top_p=0.7,
        max_tokens=8192,
        extra_body={"chat_template_kwargs": {"thinking": False}},
        stream=stream,
    )

    if stream:
        for chunk in completion:
            if getattr(chunk, "choices", None) and chunk.choices[0].delta.content is not None:
                print(chunk.choices[0].delta.content, end="")
        print()
    else:
        return completion.choices[0].message.content
```

```python
send_user_request("What is the capital of France?")
```
